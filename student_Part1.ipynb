## Merge Data

```python
# Merge the Data files to be used
df_Merged = pd.merge(df_tnMovieBudgets, df_imdbTitleBasics, left_on="movie", right_on="original_title" )
df_Merged = pd.merge(df_Merged, df_imdbTitleRatings, on="tconst")
df_Merged = pd.merge(df_Merged, df_imdbTitlePrincipals, on="tconst")
df_Merged = pd.merge(df_Merged, df_imdbNameBasics, on="nconst")
df_Merged.info()
df_Merged.tail()
```

## Clean Data

```python
# Remove the unnecessary columns
drop_Columns = ['id', 'primary_title', 'original_title', 'runtime_minutes', 'ordering',\
                'job', 'start_year', 'birth_year', 'primary_profession', 'known_for_titles' ]
df_Merged.drop(columns = drop_Columns, inplace=True)
df_Merged.info()
```

```python
# Remove rows with $0 worldwide_gross
df_Merged.drop(df_Merged[df_Merged.worldwide_gross == '$0'].index, inplace=True)
df_Merged.info()
```

```python
df_Merged[df_Merged.production_budget == '$0']
```

```python
# Change the currency columns to float
df_Merged['production_budget'] = df_Merged['production_budget'].str.replace(",", "")
df_Merged['production_budget'] = df_Merged['production_budget'].str.replace("$", "")
df_Merged['production_budget'] = df_Merged['production_budget'].astype(float)

df_Merged['domestic_gross'] = df_Merged['domestic_gross'].str.replace(",", "")
df_Merged['domestic_gross'] = df_Merged['domestic_gross'].str.replace("$", "")
df_Merged['domestic_gross'] = df_Merged['domestic_gross'].astype(float)

df_Merged['worldwide_gross'] = df_Merged['worldwide_gross'].str.replace(",", "")
df_Merged['worldwide_gross'] = df_Merged['worldwide_gross'].str.replace("$", "")
df_Merged['worldwide_gross'] = df_Merged['worldwide_gross'].astype(float)
```

```python
df_Merged.info()
df_Merged.head()
```

```python
# Seperate genres
df_Merged['genres'] = df_Merged['genres'].str.split(',')
df_Merged = df_Merged.explode('genres')
```

```python
# Slice the release year from the release date
df_Merged['release_year'] = df_Merged['release_date'].str.slice(start=-4).astype(int)
```

```python
df_Merged.info()
df_Merged.head()
```

## Create Profit columns

```python
#Two types of profits
df_Merged['profit_gross'] = df_Merged['worldwide_gross'] - df_Merged['production_budget']
df_Merged['profit_rate'] = df_Merged['worldwide_gross'] / df_Merged['production_budget']
df_Merged.info()
df_Merged.head()
```

```python
#Sort values based on production_budget
df_Merged.sort_values(by='production_budget', ascending=False, inplace=True, ignore_index=True)
```
